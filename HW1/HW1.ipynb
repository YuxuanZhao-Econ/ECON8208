## 1 (a)

$x, f(x)$ are both scalars and $f'(x)$ is analytically known, we need to find the root of $f(x) = x$. 

Define $g(x) = f(x) - x$, then we need to find the root of $g(x) = 0$.

We can use the Newton method to iteratively find the root.

### Pseudocode for Newton's method

**Input:** $f, f', x^0$, `tol`, `max_iter`  

**Output:** $x^*$, numbers of iteration, convergence status

**Algorithm:**

1. Initialize $x = x^0$.
2. For $i = 1, 2, ...$, max_iter:
   - Compute $g(x^i) = f(x^i) - x^i$.
   - Compute $g'(x^i) = f'(x^i) - 1$.
   - Update
     $$
     x^{i+1} = x^i - \frac{g(x^i)}{g'(x^i)}.
     $$
    
- If $|g(x^{i+1})| < \text{tol}$, set $x^* = x^{i+1}$ and stop, return $x^*$, $i$, `true`.
- If $i =$ max_iter and $|g(x^{i+1})| > \text{tol}$, return $x^* = x^{i+1}$, $i$, `false` (with a warning about convergence)



```julia
# -------------------------------------------------------
# (a) x, f(x) are scalars and f'(x) is analytically known
# Newton method for g(x) = f(x) - x = 0
# -------------------------------------------------------
function fixed_point_a(f, fp, x0; tol=1e-10, max_iter=1000)
    x = Float64(x0)

    for iter in 1:max_iter
        g = f(x) - x
        gp = fp(x) - 1.0

        if abs(g) < tol
                println("Converged to fixed point: $x in $iter iterations.")
            return x, iter, true
        end

        x_new = x - g / gp # Newton update
        x = x_new
    end

    println("Warning: Did not converge to a fixed point within $max_iter iterations.")
    return x, max_iter, false
end
```

## 1 (b)

$x, f(x)$ are both scalars and $f'(x)$ is not known

Define $g(x) = f(x) - x$, then we need to find the root of $g(x) = 0$.

We can use the Newton method to iteratively find the root. But since $f'(x)$ is not known, we need to use a numerical differentiation function to approximate $f'(x)$.

### Numerical differentiation
To approximate $f'(x)$, we can use the central difference formula:
$$
f'(x) \approx \frac{f(x + h) - f(x - h)}{2h},
$$
where $h$ is a small step size.

### Pseudocode for numerical differentiation

**Input:** $f, x, h$

**Output:** $f'(x)$

**Algorithm:**

1. Compute $f(x + h)$ and $f(x - h)$.
2. Compute the central difference: $f'(x) \approx \frac{f(x + h) - f(x - h)}{2h}$.
3. Return $f'(x)$.

### Pseudocode for Newton's method

**Input:** $f, x^0$, `tol`, `max_iter`, h  

**Output:** $x^*$, numbers of iteration, convergence status

**Algorithm:**

1. Initialize $x = x^0$.
2. For $i = 1, 2, ...$, max_iter:
   - Compute $g(x^i) = f(x^i) - x^i$.
   - Compute $f'(x^i) = \text{numerical\_derivative}(f, x^i, h)$ using the numerical differentiation function.
   - Compute $g'(x^i) = f'(x^i) - 1$.
   - Update
     $$
     x^{i+1} = x^i - \frac{g(x^i)}{g'(x^i)}.
     $$
    
- If $|g(x^{i+1})| < \text{tol}$, set $x^* = x^{i+1}$ and stop, return $x^*$, $i$, `true`.
- If $i =$ max_iter and $|g(x^{i+1})| > \text{tol}$, return $x^* = x^{i+1}$, $i$, `false` (with a warning about convergence)


```julia
# -------------------------------------------------------
# (b) x, f(x) are scalars and f'(x) is not known
# Define a numerical derivative function to approximate f'(x)
# Newton method for g(x) = f(x) - x = 0
# -------------------------------------------------------

# numerical differentiation using central difference
function numerical_derivative(f, x; h=1e-6)
    return (f(x + h) - f(x - h)) / (2.0 * h)
end

# fixed point solver for scalar case when f'(x) is unknown
function fixed_point_b(f, x0; tol=1e-10, max_iter=1000, h=1e-6)
    x = Float64(x0)

    for i in 1:max_iter
        g = f(x) - x
        fp_approx = numerical_derivative(f, x; h=h)
        gp = fp_approx - 1.0

        if abs(g) < tol
                println("Converged to fixed point: $x in $i iterations.")
            return x, i, true
        end

        x_new = x - g / gp # Newton update

        x = x_new
    end

    println("Warning: Did not converge to a fixed point within $max_iter iterations.")
    return x, max_iter, false
end
```

## 1 (c)

$x, f(x)$ are both vectors and derivatives $\partial f_i(x)/\partial x_j$ are analytically known

Define $g(x) = f(x) - x$, then we need to find the root of $g(x) = 0$.

We can use the Newton method to iteratively find the root.

### Pseudocode for Newton's method

**Input:** $f, f', x^0$, `tol`, `max_iter`  

**Output:** $x^*$, numbers of iteration, convergence status

**Algorithm:**

1. Initialize $x = x^0$.
2. For $i = 1, 2, ...$, max_iter:
   - Compute $g(x^i) = f(x^i) - x^i$.
   - Compute $g'(x^i) = f'(x^i) - 1$.
   - Update
     $$
     x^{i+1} = x^i - (g'(x^i))^{-1} g(x^i).
     $$
    
- If $|g(x^{i+1})| < \text{tol}$, set $x^* = x^{i+1}$ and stop, return $x^*$, $i$, `true`.
- If $i =$ max_iter and $|g(x^{i+1})| > \text{tol}$, return $x^* = x^{i+1}$, $i$, `false` (with a warning about convergence)




```julia
# -------------------------------------------------------
# (c) x, f(x) are vectors and the derivatives of f are analytically known
# Newton method for g(x) = f(x) - x = 0
# -------------------------------------------------------

# fixed point solver for vector case when the Jacobian is analytically known
function fixed_point_c(f, fp, x0; tol=1e-10, max_iter=1000)
    x = Vector{Float64}(x0)
    n = length(x)
    I_n = Matrix{Float64}(I, n, n)

    for i in 1:max_iter
        g = f(x) - x

        if norm(g, Inf) < tol
                println("Converged to fixed point: $x in $i iterations.")
            return x, i, true
        end

        # Newton update: x_new = x - (gp)^{-1} * g
        gp = fp(x) - I_n # Jacobian of g is fp - I
        dx = gp \ g 
        x_new = x - dx

        x = x_new
    end

    println("Warning: Did not converge to a fixed point within $max_iter iterations.")
    return x, max_iter, false
end
```

```julia
# ----------------------------------------------
# Test the function with a simple example
# f(x,y) = [cos(x)+sin(y); x^2 + y^3]
# fp(x,y) = [ -sin(x)  cos(y) ; 2x  3y^2 ]
# ----------------------------------------------

f(x) = [
    cos(x[1]) + sin(x[2]);
    x[1]^2 + x[2]^3
]

fp(x) = [
    -sin(x[1])   cos(x[2]);
    2.0 * x[1]   3.0 * x[2]^2
]

x0 = [0.5, 0.5]

x_star, n_iter, converged = fixed_point_c(f, fp, x0)
```

## 1 (d)

$x$ and $f(x)$ are both vectors, and the derivatives $\partial f_i(x) / \partial x_j$ are not known.

Define $g(x) = f(x) - x$, then we need to find the root of $g(x) = 0$.

We can use the Newton method to iteratively find the root. Since the derivatives are not analytically available, we approximate the Jacobian matrix numerically.

### Numerical differentiation

To approximate the partial derivatives, we can use the central difference formula. For each component $f_i(x)$ and variable $x_j$,
$$
\frac{\partial f_i(x)}{\partial x_j}
\approx
\frac{f_i(x + h e_j) - f_i(x - h e_j)}{2h},
$$
where $h$ is a small step size and $e_j$ is the $j$ th unit vector.

Thus, the Jacobian matrix $J_f(x)$ can be approximated numerically, and
$$
J_g(x) = J_f(x) - I.
$$

### Pseudocode for numerical Jacobian approximation

**Input:** $f, x, h$

**Output:** $J_f(x)$

**Algorithm:**

1. For each variable $x_j$, construct the unit vector $e_j$.
2. Compute
   $$
   \frac{\partial f_i(x)}{\partial x_j}
   \approx
   \frac{f_i(x + h e_j) - f_i(x - h e_j)}{2h}
   $$
   for all $i = 1, \dots, n$.
3. Form the Jacobian matrix $J_f(x)$.
4. Return $J_f(x)$.

### Pseudocode for Newton's method

**Input:** $f, x^0$, `tol`, `max_iter`, $h$

**Output:** $x^*$, number of iterations, convergence status

**Algorithm:**

1. Initialize $x = x^0$.
2. For $i = 1, 2, ...$, max_iter:
   - Compute $g(x^i) = f(x^i) - x^i$.
   - Compute $g'(x^i) = numerical_derivative(f, x^i, h) - I$.
   - Update
     $$
     x^{i+1} = x^i - (g'(x^i))^{-1} g(x^i).
     $$
    
- If $|g(x^{i+1})| < \text{tol}$, set $x^* = x^{i+1}$ and stop, return $x^*$, $i$, `true`.
- If $i =$ max_iter and $|g(x^{i+1})| > \text{tol}$, return $x^* = x^{i+1}$, $i$, `false` (with a warning about convergence)


```julia
# -------------------------------------------------------
# Numerical Jacobian using central difference
# Input: f, x, h
# Output: J_f(x)
# -------------------------------------------------------
function numerical_jacobian(f, x; h=1e-6)
    x = Vector{Float64}(x)
    n = length(x)
    J = zeros(n, n)

    for j in 1:n
        e = zeros(n)
        e[j] = 1.0

        f_plus = f(x + h * e)
        f_minus = f(x - h * e)

        J[:, j] = (f_plus - f_minus) / (2.0 * h)
    end

    return J
end


# -------------------------------------------------------
# (d) x, f(x) are both vectors and derivatives are not known
# Newton method for g(x) = f(x) - x = 0
# -------------------------------------------------------
function fixed_point_d(f, x0; tol=1e-10, max_iter=1000, h=1e-6)
    x = Vector{Float64}(x0)
    n = length(x)
    I_n = Matrix{Float64}(I, n, n)

    for i in 1:max_iter
        g = f(x) - x

        if norm(g, Inf) < tol
                println("Converged to fixed point: $x in $i iterations.")
            return x, i, true
        end

        gp = numerical_jacobian(f, x; h=h) - I_n
        dx = gp \ g
        x_new = x - dx

        x = x_new
    end

    println("Warning: Did not converge to a fixed point within $max_iter iterations.")
    return x, max_iter, false
end
```

## 2 (a)

Given user-specified values for $c$, $z$, $k$, $u(\cdot)$, and $f(\cdot)$, we want to compute the fixed point $\ell^*$ of the intratemporal condition in the standard growth model:
$$
0 = \frac{u_{\ell}(c,\ell)}{u_c(c,\ell)} - z f_h(k,h),
\qquad \ell = 1 - h.
$$

Since $\ell = 1-h$, we can rewrite the condition as a scalar root-finding problem in $\ell$:
$$
g(\ell) = \frac{u_{\ell}(c,\ell)}{u_c(c,\ell)} - z f_h\bigl(k, 1-\ell\bigr) = 0.
$$

Thus, solving for the fixed point $\ell^*$ is equivalent to finding the root of $g(\ell)$.

Because $\ell$ is a scalar, we can use the scalar methods developed in Part 1.  
If the derivative of $g(\ell)$ is not analytically available, we can use the numerical-derivative Newton method from Part 1(b).


### Pseudocode for numerical differentiation

We only have 
$$
u(c,\ell), f(k,h), g(\ell)
$$
but in the root-finding algorithm, we need to compute 
$$
u_\ell(c,\ell), u_c(c,\ell), f_h(k,h), g'(\ell)
$$
We can use the central difference formula to approximate them, for any function $w(x)$,
$$
w'(x) \approx \frac{w(x + h) - w(x - h)}{2h}.
$$
We define a numerical differentiation function that takes for vector-valued functions and returns the Jacobian matrix. We just need to call it now.




### Pseudocode

**Input:** $c$, $z$, $k$, $u(c,\ell)$, $f(k,h)$, initial guess $\ell^0$, tol, max\_iter, and $h$ for numerical differentiation.

 

**Output:** $\ell^*$, number of iterations, convergence status

**Algorithm:**

1. Define the labor choice as
   $$
   h = 1 - \ell.
   $$
2. Given the inputs, we can compute $u_\ell(c,\ell)$, $u_c(c,\ell)$, and $f_h(k,h)$ using numerical differentiation defined in 1(d).
3. Define the scalar function
   $$
   g(\ell) = \frac{u_{\ell}(c,\ell)}{u_c(c,\ell)} - z f_h(k, 1-\ell).
   $$
4. Choose an initial guess $\ell^0$. Then we apply the scalar root-finding method from 1 (d) to solve
   $$
   g(\ell)=0.
   $$
   1. For For $i = 1, 2, ...$, max_iter:
      - Compute $g(\ell^i)$.
      - Compute $g'(\ell^i)$ using numerical differentiation.
      - Update
        $$
        \ell^{i+1} = \ell^i - \frac{g(\ell^i)}{g'(\ell^i)}.
        $$
      - then check that the candidate value of $\ell^{i+1}$ is feasible:
      $$
      0 \le \ell^{i+1} \le 1,
      $$
      If the candidate $\ell^{i+1}$ is outside the feasible interval, return $\ell^* = \ell^{i+1}$, number of iterations, `false`, and print a warning.
      - If
      $$
      |g(\ell^{i+1})| < \text{tol},
      $$
      return $\ell^* = \ell^{i+1}$, the iteration count, and `true`.
      - If $|g(\ell^{i+1})| > \text{tol}$ and $i+1 = max\_iter$, return $\ell^* = \ell^{i+1}$, number of iterations, and `false`, together with a warning about non-convergence.


```julia
# -------------------------------------------------------
# 2(a): solve for l* in
# g(l) = u_l(c,l) / u_c(c,l) - z * f_h(k, 1-l) = 0
# Input: c, z, k, u(c,l), f(k,h), initial guess l0, tol, max_iter, and h for numerical differentiation.
# Output: l*, number of iterations, convergence status
# -------------------------------------------------------
function solve_l_star(c, z, k, u, f, l0; tol=1e-10, max_iter=1000, h=1e-6)
    l = Float64(l0)

    # partial derivative u_c(c,l)
    function u_c(cval, lval)
        w(x) = u(x, lval)
        return numerical_derivative(w, cval; h=h)
    end

    # partial derivative u_l(c,l)
    function u_l(cval, lval)
        w(x) = u(cval, x)
        return numerical_derivative(w, lval; h=h)
    end

    # partial derivative f_h(k,h)
    function f_h(kval, hval)
        w(x) = f(kval, x)
        return numerical_derivative(w, hval; h=h)
    end

    # g(l) = u_l/u_c - z*f_h
    function g(lval)
        hval = 1.0 - lval

        uc_val = u_c(c, lval)
        ul_val = u_l(c, lval)
        fh_val = f_h(k, hval)

        if abs(uc_val) < eps(Float64)
            println("Warning: u_c is too close to zero, g(l) is not well-defined.")
            return NaN
        end

        return ul_val / uc_val - z * fh_val
    end

    # Initial check
    
    if l0 < 0.0 || l0 > 1.0
        println("Warning: initial guess for l is outside [0,1].")
        return l0, 0, false
    end

    # check if g(l) is close to zero
    if abs(g(l0)) < tol
        println("Initial guess for l is already a solution.")
        return l0, 0, true
    end


    for i in 1:max_iter

        g_val = g(l)

        # check if g(l) is finite
        if !isfinite(g_val)
            println("Warning: g(l) is not finite at the current iterate.")
            return l, i, false
        end

        gp_val = numerical_derivative(g, l; h=h)

        # check if g'(l) is finite
        if !isfinite(gp_val)
            println("Warning: g'(l) is not finite at the current iterate.")
            return l, i, false
        end

        # check if g'(l) is too close to zero
        # if g'(l) is close to zero, we can not multiply by 1/g'(l) in the Newton update
        if abs(gp_val) < eps(Float64)
            println("Warning: g'(l) is too close to zero, cannot perform Newton update.")
            return l, i, false
        end

        # Newton update: l_new = l - g(l)/g'(l)
        l_new = l - g_val / gp_val


        if l_new < 0.0 || l_new > 1.0
            println("Warning: l is outside the feasible interval [0,1].")
            return l_new, i, false
        end

        g_val_new = g(l_new)

        # check if g(l) is close to zero
        if abs(g_val_new) < tol
            println("Converged to solution: $l_new in $(i+1) iterations.")
            return l_new, i+1, true
        end


        l = l_new
    end

    println("Warning: method did not converge within max_iter.")
    return l, max_iter, false
end
```

### Robustness considerations

To make the code robust, we include the following checks:

1. Check that the initial guess $\ell^0$ is feasible, i.e.
   $$
   0 \le \ell^0 \le 1.
   $$
   If not, stop immediately and return `false`.

2. At each iteration, check that the current iterate $\ell^i$ remains in the feasible interval
   $$
   0 \le \ell^i \le 1.
   $$
   If not, stop the algorithm and return a warning about infeasibility.

3. Before evaluating
   $$
   \frac{u_\ell(c,\ell)}{u_c(c,\ell)},
   $$
   check that $u_c(c,\ell)$ is not too close to zero. Otherwise, the ratio is not numerically stable.

4. Check that the function value $g(\ell^i)$ is finite. If $g(\ell^i)$ is `NaN` or `Inf`, stop and return `false`.

5. Check that the numerical derivative $g'(\ell^i)$ is finite. If it is not finite, stop and return `false`.

6. Check that the numerical derivative $g'(\ell^i)$ is not too close to zero. Otherwise, the Newton update
   $$
   \ell^{i+1} = \ell^i - \frac{g(\ell^i)}{g'(\ell^i)}
   $$
   is not well-defined.

7. If the method does not converge within `max_iter`, return the last iterate together with a warning about non-convergence.


```julia
u(c, l) = log(c) + log(l)
f(k, h) = k^0.3 * h^0.7

c = 1.0
z = 1.0
k = 1.0
l0 = 0.5

l_star, n_iter, converged = solve_l_star(c, z, k, u, f, l0)

println("l_star = ", l_star)
println("iterations = ", n_iter)
println("converged = ", converged)
println("h_star = ", 1.0 - l_star)
```

## 2 (b)

### Shopping time model

Given specified $w$, $y$, $u(\cdot)$, and $p(\cdot)$, households maximize:
$$
\begin{aligned}
&\max_{h, s} u(c, \ell), \quad \text{s.t.} \\
&p(s) c = w h + y, \\
&\ell = 1 - h - s
\end{aligned}
$$

### First-order conditions

The Lagrangian is
$$
\mathcal{L} = u(c,\ell) + \lambda (wh + y - p(s)c) + \mu (1 - h - s - \ell).
$$

Taking derivatives with respect to each choice variable:

$$
\begin{aligned}
&\frac{\partial \mathcal{L}}{\partial c} = u_c(c,\ell) - \lambda p(s) = 0 \\
&\frac{\partial \mathcal{L}}{\partial \ell} = u_\ell(c,\ell) - \mu = 0 \\
&\frac{\partial \mathcal{L}}{\partial h} = \lambda w - \mu = 0 \\
&\frac{\partial \mathcal{L}}{\partial s} = -\lambda p'(s)c - \mu = 0 \\
\end{aligned}
$$

Rearranging, we have
$$
\begin{aligned}
u_c(c,\ell)\frac{w}{p(s)} - u_\ell(c,\ell) = 0, \\
-u_c(c,\ell)\frac{p'(s)c}{p(s)} - u_\ell(c,\ell) = 0, \\
\end{aligned}
$$

Plugging in the budget constraint, we can rewrite the system as
$$
\begin{aligned}
&g_1(h,s) = u_c\left(\frac{wh+y}{p(s)}, 1-h-s\right)\frac{w}{p(s)} - u_\ell\left(\frac{wh+y}{p(s)}, 1-h-s\right) = 0, \\
&g_2(h,s) = -u_c\left(\frac{wh+y}{p(s)}, 1-h-s\right)\frac{p'(s)(wh+y)}{p(s)^2} - u_\ell\left(\frac{wh+y}{p(s)}, 1-h-s\right) = 0. \\
\end{aligned}
$$

We can solve for the $(h^*, s^*)$ by finding the root of the system of equations $g(h,s) = 0$.



### Root-finding problem

Given user-specified values for $w$, $y$, $u(\cdot)$, and $p(\cdot)$, we want to compute the $(h^*, s^*)$ such that $G(h,s) = 0$.


We define the vector-valued function
$$
G(h,s) =
\begin{bmatrix}
g_1(h,s) \\
g_2(h,s)
\end{bmatrix},
$$

Because $(h,s)$ is a vector and the derivatives are not analytically available, we use the numerical-Jacobian Newton method from Part 1(d).



### Numerical differentiation

We only have
$$
u(c,\ell), \qquad p(s),
$$
but in the root-finding algorithm we need
$$
u_c(c,\ell), \qquad u_\ell(c,\ell), \qquad p'(s),
$$
as well as the Jacobian of $G(h,s)$.

We approximate these derivatives numerically using the central difference formula:
$$
w'(x) \approx \frac{w(x+h)-w(x-h)}{2h}.
$$



### Pseudocode

**Input:** $w$, $y$, $u(c,\ell)$, $p(s)$, initial guess $(h^0,s^0)$, `tol`, `max_iter`, and $h$ for numerical differentiation.

**Output:** $(h^*, s^*)$, number of iterations, convergence status

**Algorithm:**

1. Define
   $$
   c(h,s) = \frac{wh+y}{p(s)},
   \qquad
   \ell(h,s) = 1-h-s.
   $$
2. Using numerical differentiation, compute:
   - $u_c(c,\ell)$,
   - $u_\ell(c,\ell)$,
   - $p'(s)$.
3. Define the vector-valued function
   $$
   G(h,s) =
   \begin{bmatrix}
   u_c(c,\ell)\frac{w}{p(s)} - u_\ell(c,\ell) \\
   u_c(c,\ell)\left(-\frac{(wh+y)p'(s)}{p(s)^2}\right) - u_\ell(c,\ell)
   \end{bmatrix}.
   $$
4. Choose an initial guess
   $$
   x^0 =
   \begin{bmatrix}
   h^0 \\
   s^0
   \end{bmatrix}.
   $$
5. Apply the vector root-finding method from Part 1(d) to solve
   $$
   G(x)=0.
   $$
   1. For $i = 1, 2, \ldots, \text{max\_iter}$:
      - Compute $G(x^i)$.
      - Compute the numerical Jacobian $J_G(x^i)$.
      - Update
        $$
        x^{i+1} = x^i - J_G(x^i)^{-1} G(x^i).
        $$
      - Let
        $$
        x^{i+1} =
        \begin{bmatrix}
        h^{i+1} \\
        s^{i+1}
        \end{bmatrix}.
        $$
      - Check feasibility:
        $$
        h^{i+1} \ge 0,\qquad s^{i+1} \ge 0,\qquad h^{i+1}+s^{i+1} \le 1.
        $$
      - Also check that
        $$
        p(s^{i+1}) > 0.
        $$
      - If any of these conditions fail, return the previous iterate, the iteration count, and `false`, together with a warning.
      - If
        $$
        \|G(x^{i+1})\|_\infty < \text{tol},
        $$
        return $(h^*, s^*) = x^{i+1}$, the iteration count, and `true`.
      - If $i+1 = \text{max\_iter}$ and $\|G(x^{i+1})\|_\infty > \text{tol}$, return $(h^*, s^*) = x^{i+1}$, the iteration count, and `false`, together with a warning about non-convergence.



```julia
# -------------------------------------------------------
# 2(b): solve for (h*, s*) in the system of equations G(h,s) = 0
# where G(h,s) = [g1(h,s); g2(h,s)] is defined by the first-order conditions of the consumer's problem.
# Input: w, y, u(c,l), p(s), initial guess x0 = (h0, s0), tol, max_iter, and h for numerical differentiation.
# Output: (h*, s*), number of iterations, convergence status
# -------------------------------------------------------



function solve_hs_star(w, y, u, p, x0; tol=1e-10, max_iter=1000, h=1e-6)
    x = Vector{Float64}(x0)

    # feasibility check
    function is_feasible(x)
        h_work = x[1]
        s_shop = x[2]
        l = 1.0 - h_work - s_shop

        if h_work < 0.0 || s_shop < 0.0 || l < 0.0
            println("Warning: (h, s) is infeasible because h_work = $h_work, s_shop = $s_shop, l = $l.")
            return false
        end

        p_val = p(s_shop)
        if !isfinite(p_val) || p_val <= 0.0
            println("Warning: price p(s) is not well-defined or non-positive at s = $s_shop.")
            return false
        end

        return true
    end


    # Compute the necessary derivatives for G(h,s)

    function u_c(cval, lval)
        w_tmp(z) = u(z, lval)
        return numerical_derivative(w_tmp, cval; h=h)
    end

    function u_l(cval, lval)
        w_tmp(z) = u(cval, z)
        return numerical_derivative(w_tmp, lval; h=h)
    end

    function p_prime(sval)
        return numerical_derivative(p, sval; h=h)
    end

    function c_of(h_work, s_shop)
        return (w * h_work + y) / p(s_shop)
    end

    function l_of(h_work, s_shop)
        return 1.0 - h_work - s_shop
    end

    # define the system G(h,s) = [g1(h,s); g2(h,s)]
    function G(xvec)
        h_work = xvec[1]
        s_shop = xvec[2]

        c_val = c_of(h_work, s_shop)
        l_val = l_of(h_work, s_shop)

        uc_val = u_c(c_val, l_val)
        ul_val = u_l(c_val, l_val)
        p_val = p(s_shop)
        pp_val = p_prime(s_shop)

        g1 = uc_val * (w / p_val) - ul_val
        g2 = uc_val * (-(w * h_work + y) * pp_val / (p_val^2)) - ul_val

        return [g1; g2]
    end


    # initial check

    if !is_feasible(x0)
        return x0, 0, false
    end

    Gx0 = G(x0)

    if norm(Gx0, Inf) < tol
        println("Initial guess for (h, s) is already a solution.")
        return x0, 0, true
    end


    # main iteration loop
    # use newton's method to solve G(h,s) = 0
    for i in 1:max_iter

        Gx = G(x)

        # check if G(x) is well-defined
        if !all(isfinite.(Gx))
            println("Warning: G(x) is not finite at the current iterate.")
            return x, i, false
        end


        JG = numerical_jacobian(G, x; h=h)

        # check if numerical Jacobian is well-defined
        if !all(isfinite.(JG))
            println("Warning: numerical Jacobian is not finite at the current iterate.")
            return x, i, false
        end

        # check if numerical Jacobian is singular
        # if the Jacobian is singular, we cannot perform the Newton update by multiplying JG^{-1}
        if rank(JG) < length(x)
            println("Warning: numerical Jacobian is singular at the current iterate.")
            return x, i, false
        end

        # check if numerical Jacobian is close to zero
        if norm(JG, Inf) < eps(Float64)
            println("Warning: numerical Jacobian is too close to zero, cannot perform Newton update.")
            return x, i, false
        end

        dx = JG \ Gx
        x_new = x - dx

        # check feasibility of current x
        if !is_feasible(x_new)
            return x_new, i, false
        end

        Gx = G(x_new)

        if norm(Gx, Inf) < tol
            println("Converged to solution: $x_new in $i iterations.")
            return x_new, i, true
        end


        x = x_new
    end

    println("Warning: method did not converge within max_iter.")
    return x, max_iter, false
end
```

### Robustness considerations

To make the code robust, we include the following checks:

1. Check that the initial guess
   $$
   x^0 =
   \begin{bmatrix}
   h^0 \\
   s^0
   \end{bmatrix}
   $$
   is feasible. In particular, require
   $$
   h^0 \ge 0,\qquad s^0 \ge 0,\qquad 1-h^0-s^0 \ge 0.
   $$
   If not, stop immediately and return `false`.

2. At each iteration, check that the current iterate
   $$
   x^i =
   \begin{bmatrix}
   h^i \\
   s^i
   \end{bmatrix}
   $$
   remains feasible, i.e.
   $$
   h^i \ge 0,\qquad s^i \ge 0,\qquad 1-h^i-s^i \ge 0.
   $$

3. Check that the price function is well-defined and strictly positive:
   $$
   p(s^i) > 0.
   $$
   If $p(s^i)$ is not finite or is non-positive, stop and return a warning.

4. Check that the system value
   $$
   G(x^i)
   $$
   is finite. If any component of $G(x^i)$ is `NaN` or `Inf`, stop and return `false`.

5. Check that the numerical Jacobian
   $$
   J_G(x^i)
   $$
   is finite. If any entry is `NaN` or `Inf`, stop and return `false`.

6. Check that the numerical Jacobian is not singular. If
   $$
   \mathrm{rank}(J_G(x^i)) < \text{length}(x^i),
   $$
   then the Newton step cannot be computed reliably, so stop and return a warning.

7. Check that the numerical Jacobian is not too close to zero. If its norm is too small, the Newton update may be unstable.

8. If the method does not converge within `max_iter`, return the last iterate together with a warning about non-convergence.


```julia
# -----------------------------------------------
# Test the function with the shopping-time model
# u(c, l) = log(c) + log(l)
# p(s) = exp(-2.0 * s)
# -----------------------------------------------


u(c, l) = log(c) + log(l)
p(s) = exp(-2.0 * s)

w = 1.0
y = 0.2
x0 = [0.2, 0.3]

x_star, n_iter, converged = solve_hs_star(w, y, u, p, x0)

println("x_star = ", x_star)
println("h_star = ", x_star[1])
println("s_star = ", x_star[2])
println("iterations = ", n_iter)
println("converged = ", converged)
```
