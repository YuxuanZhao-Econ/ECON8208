# HW 6

Compute equilibria of the following growth model with fiscal shocks:
$$
\begin{aligned}
\max_{\{c_t,x_t,\ell_t\}} \quad
& E_0 \sum_{t=0}^{\infty} \beta^t
\left\{
\frac{(c_t \ell_t^{\psi})^{1-\sigma}}{1-\sigma}
\right\} N_t \\
\text{s.t.} \quad
& c_t + x_t
= r_t k_t + w_t h_t + \kappa_t
- \tau_t^c c_t
- \tau_t^h w_t h_t
- \tau_t^p (r_t k_t - \delta k_t) \\
& \hspace{2.45cm}
- \tau_t^d \Bigl(r_t k_t - x_t - \tau_t^p (r_t k_t - \delta k_t)\Bigr), \\
& N_{t+1} k_{t+1} = \bigl[(1-\delta)k_t + x_t\bigr]N_t, \\
& h_t + \ell_t = 1, \\
& S_t = P_0 + P S_{t-1} + Q \epsilon_t, \\
& S_t = [\log z_t,\tau_t^c,\tau_t^h,\tau_t^d,\tau_t^p,\log g_t]', \\
& N_t = (1+\gamma_n)^t, \\
& c_t, x_t \ge 0 \quad \text{in all states.}
\end{aligned}
$$

The production technology is
$$
Y_t = K_t^\theta \bigl(Z_t H_t\bigr)^{1-\theta},
\qquad
Z_t = (1+\gamma_z)^t z_t,
$$
and factor prices satisfy
$$
r_t = \theta K_t^{\theta-1}(Z_t H_t)^{1-\theta},
\qquad
w_t = (1-\theta)K_t^\theta Z_t^{1-\theta} H_t^{-\theta}.
$$

In equilibrium, we impose the resource constraint and the government budget constraint:
$$
\begin{aligned}
C_t + X_t + G_t &= Y_t, \\
G_t + N_t \kappa_t
&=
\tau_t^c C_t
+
\tau_t^h w_t H_t
+
\tau_t^p (r_t K_t - \delta K_t)
+
\tau_t^d \Bigl(r_t K_t - X_t - \tau_t^p (r_t K_t - \delta K_t)\Bigr).
\end{aligned}
$$

This model also has a balanced growth path because productivity grows at the labor-augmenting rate $\gamma_z$. Hence, before solving the equilibrium we detrend the variables that grow over time.

Define the detrended variables as
$$
\tilde c_t = \frac{c_t}{(1+\gamma_z)^t}, \qquad
\tilde x_t = \frac{x_t}{(1+\gamma_z)^t}, \qquad
\tilde k_t = \frac{k_t}{(1+\gamma_z)^t}, \\
\tilde \kappa_t = \frac{\kappa_t}{(1+\gamma_z)^t}, \qquad
\tilde g_t = \frac{g_t}{(1+\gamma_z)^t}, \qquad
\tilde G_t = \frac{G_t}{N_t (1+\gamma_z)^t}, \qquad
\tilde h_t = h_t, \qquad \tilde \ell_t = \ell_t, \\
\tilde Y_t = \frac{Y_t}{N_t (1+\gamma_z)^t}, \qquad
\tilde K_t = \frac{K_t}{N_t (1+\gamma_z)^t}, \qquad
\tilde H_t = \frac{H_t}{N_t}, \qquad
\tilde r_t = r_t, \qquad
\tilde w_t = \frac{w_t}{(1+\gamma_z)^t}. \\
\tilde X_t = \frac{X_t}{N_t (1+\gamma_z)^t}, \qquad
\tilde C_t = \frac{C_t}{N_t (1+\gamma_z)^t}.
$$

Then period utility becomes
$$
N_t \frac{(c_t \ell_t^\psi)^{1-\sigma}}{1-\sigma}
=
(1+\gamma_n)^t (1+\gamma_z)^{t(1-\sigma)}
\frac{(\tilde c_t \tilde \ell_t^\psi)^{1-\sigma}}{1-\sigma}.
$$

Therefore, after detrending, the effective discount factor is
$$
\tilde \beta
=
\beta (1+\gamma_n)(1+\gamma_z)^{1-\sigma}.
$$

The detrended capital accumulation equation is
$$
\tilde k_{t+1}
=
\frac{(1-\delta)\tilde k_t + \tilde x_t}{(1+\gamma_n)(1+\gamma_z)}
$$


Therefore, the detrended household problem can be written as
$$
\begin{aligned}
\max_{\{\tilde c_t,\tilde x_t,\tilde \ell_t\}} \quad
& E_0 \sum_{t=0}^{\infty} \tilde \beta^t
\left\{
\frac{(\tilde c_t \tilde \ell_t^{\psi})^{1-\sigma}}{1-\sigma}
\right\} \\
\text{s.t.} \quad
& \tilde c_t + \tilde x_t
=
\tilde r_t \tilde k_t + \tilde w_t \tilde h_t + \tilde \kappa_t
- \tau_t^c \tilde c_t
- \tau_t^h \tilde w_t \tilde h_t
- \tau_t^p (\tilde r_t \tilde k_t - \delta \tilde k_t) \\
& \hspace{2.2cm}
- \tau_t^d \Bigl(\tilde r_t \tilde k_t - \tilde x_t - \tau_t^p (\tilde r_t \tilde k_t - \delta \tilde k_t)\Bigr), \\
& \tilde k_{t+1}
=
\frac{(1-\delta)\tilde k_t + \tilde x_t}{(1+\gamma_n)(1+\gamma_z)}, \\
& \tilde h_t + \tilde \ell_t = 1, \\
& S_t = P_0 + P S_{t-1} + Q \epsilon_t, \\
& S_t = [\log z_t,\tau_t^c,\tau_t^h,\tau_t^d,\tau_t^p,\log \tilde g_t]', \\
& \tilde c_t,\tilde x_t \ge 0 \quad \text{in all states.}
\end{aligned}
$$

We have
$$
\tilde Y_t = \frac{Y_t}{N_t (1+\gamma_z)^t}, \qquad
\tilde K_t = \frac{K_t}{N_t (1+\gamma_z)^t}, \qquad
\tilde H_t = \frac{H_t}{N_t}
$$

The detrended production function and factor prices are
$$
\tilde Y_t = \tilde K_t^\theta (z_t \tilde H_t)^{1-\theta}, \qquad
\tilde r_t = \theta \tilde K_t^{\theta-1}(z_t \tilde H_t)^{1-\theta},
\qquad
\tilde w_t = (1-\theta)\tilde K_t^\theta z_t^{1-\theta}\tilde H_t^{-\theta}.
$$


Hence, the detrended market clearing condition and government budget constraint can be written as
$$
\begin{aligned}
\tilde K_t = \tilde k_t, \quad
\tilde G_t = \tilde g_t, \quad
\tilde C_t &= \tilde c_t, \quad
\tilde X_t = \tilde x_t, \quad
\tilde H_t = \tilde h_t, \\
\tilde K_{t+1} &= \frac{(1-\delta)\tilde K_t + \tilde X_t}{(1+\gamma_n)(1+\gamma_z)}, \\
\tilde C_t + \tilde X_t + \tilde G_t
&= \tilde Y_t, \\
\tilde G_t + \tilde \kappa_t
&=
\tau_t^c \tilde C_t
+
\tau_t^h \tilde w_t \tilde H_t
+
\tau_t^p (\tilde r_t \tilde K_t - \delta \tilde K_t)
+
\tau_t^d \Bigl(\tilde r_t \tilde K_t - \tilde X_t - \tau_t^p (\tilde r_t \tilde K_t - \delta \tilde K_t)\Bigr).
\end{aligned}
$$


```julia
# -------------------------------------------------------
# Structural parameters
# -------------------------------------------------------
beta    = 0.96
psi     = 1.50
sigma   = 2.00
gamma_n = 0.01
gamma_z = 0.02
theta   = 0.36
delta   = 0.08

# -------------------------------------------------------
# Steady-state fiscal targets
# We normalize z_ss = 1, so log z_ss = 0
# -------------------------------------------------------
log_z_ss = 0.0
tau_c_ss = 0.05
tau_h_ss = 0.20
tau_d_ss = 0.10
tau_p_ss = 0.20
g_ss     = 0.18

# -------------------------------------------------------
# Exogenous process:
# S_t = P0 + P S_{t-1} + Q ε_t
# S_t = [log z_t, tau_c_t, tau_h_t, tau_d_t, tau_p_t, log g_t]'
#
# Assume:
#   1. Fiscal variables are uncorrelated, so P is diagonal
#   2. Shocks are uncorrelated, so Q is diagonal
# -------------------------------------------------------

# Persistence parameters
rho_z     = 0.95
rho_tau_c = 0.90
rho_tau_h = 0.90
rho_tau_d = 0.90
rho_tau_p = 0.90
rho_g     = 0.90

# Shock standard deviations
sigma_z     = 0.02
sigma_tau_c = 0.005
sigma_tau_h = 0.005
sigma_tau_d = 0.005
sigma_tau_p = 0.005
sigma_g     = 0.02

# -------------------------------------------------------
# Build steady-state exogenous vector S̄
# -------------------------------------------------------
Sbar = [
    log_z_ss,        # log z_ss
    tau_c_ss,
    tau_h_ss,
    tau_d_ss,
    tau_p_ss,
    log(g_ss)
]

# -------------------------------------------------------
# Build P and Q
# -------------------------------------------------------
P = Diagonal([
    rho_z,
    rho_tau_c,
    rho_tau_h,
    rho_tau_d,
    rho_tau_p,
    rho_g
])

Q = Diagonal([
    sigma_z,
    sigma_tau_c,
    sigma_tau_h,
    sigma_tau_d,
    sigma_tau_p,
    sigma_g
])

# -------------------------------------------------------
# Choose P0 so that E[S_t] = S_ss
# i.e. S_ss = P0 + P S_ss
# -------------------------------------------------------
P0 = (I - Matrix(P)) * Sbar

# -------------------------------------------------------
# Effective discount factor
# -------------------------------------------------------
beta_tilde = beta * (1.0 + gamma_n) * (1.0 + gamma_z)^(1.0 - sigma)

# -------------------------------------------------------
# Package parameters
# -------------------------------------------------------
params = (
    beta = beta,
    psi = psi,
    sigma = sigma,
    gamma_n = gamma_n,
    gamma_z = gamma_z,
    theta = theta,
    delta = delta,
    tau_c_ss = tau_c_ss,
    tau_h_ss = tau_h_ss,
    tau_d_ss = tau_d_ss,
    tau_p_ss = tau_p_ss,
    g_ss = g_ss,
    P0 = P0,
    P = P,
    Q = Q,
    beta_tilde = beta_tilde
)

println("P0: ", P0)
println("P: ", P)
println("Q: ", Q)
```

### Solving Steady State

We need to solve for the steady state of the model to compute the matrices that enter the LQ problem. 

#### FOC

Using $\tilde \ell_t = 1-\tilde h_t$ and letting
$$
G \equiv (1+\gamma_n)(1+\gamma_z),
$$
the detrended household problem can be written as
$$
\max E_0 \sum_{t=0}^{\infty} \tilde \beta^t
\frac{\left(\tilde c_t (1-\tilde h_t)^\psi\right)^{1-\sigma}}{1-\sigma}
$$
subject to
$$
\tilde c_t + \tilde x_t
=
\tilde r_t \tilde k_t + \tilde w_t \tilde h_t + \tilde \kappa_t
- \tau_t^c \tilde c_t
- \tau_t^h \tilde w_t \tilde h_t
- \tau_t^p (\tilde r_t \tilde k_t - \delta \tilde k_t)
- \tau_t^d \Bigl(\tilde r_t \tilde k_t - \tilde x_t - \tau_t^p (\tilde r_t \tilde k_t - \delta \tilde k_t)\Bigr),
$$
and
$$
G\tilde k_{t+1} = (1-\delta)\tilde k_t + \tilde x_t.
$$

Let $\lambda_t$ denote the multiplier on the household budget constraint and $\mu_t$ the multiplier on the capital accumulation equation. The Lagrangian is
$$
\mathcal{L}
=
E_0 \sum_{t=0}^{\infty} \tilde \beta^t
\Bigg\{
\frac{\left(\tilde c_t (1-\tilde h_t)^\psi\right)^{1-\sigma}}{1-\sigma}
+
\lambda_t \Big[
\tilde r_t \tilde k_t + \tilde w_t \tilde h_t + \tilde \kappa_t
- \tau_t^c \tilde c_t
- \tau_t^h \tilde w_t \tilde h_t
- \tau_t^p (\tilde r_t \tilde k_t - \delta \tilde k_t)
$$
$$
\qquad\qquad\qquad\qquad
- \tau_t^d \Bigl(\tilde r_t \tilde k_t - \tilde x_t - \tau_t^p (\tilde r_t \tilde k_t - \delta \tilde k_t)\Bigr)
- \tilde c_t - \tilde x_t
\Big]
+
\mu_t \Big[(1-\delta)\tilde k_t + \tilde x_t - G\tilde k_{t+1}\Big]
\Bigg\}.
$$

The first-order condition with respect to $\tilde c_t$ is
$$
\frac{\partial \mathcal{L}}{\partial \tilde c_t} = 0
\quad\Rightarrow\quad
\tilde c_t^{-\sigma}(1-\tilde h_t)^{\psi(1-\sigma)}
=
\lambda_t(1+\tau_t^c).
$$

Hence,
$$
\lambda_t
=
\frac{\tilde c_t^{-\sigma}(1-\tilde h_t)^{\psi(1-\sigma)}}{1+\tau_t^c}.
$$

The first-order condition with respect to $\tilde x_t$ is
$$
\frac{\partial \mathcal{L}}{\partial \tilde x_t} = 0
\quad\Rightarrow\quad
\lambda_t(\tau_t^d-1)+\mu_t=0.
$$

Therefore,
$$
\mu_t=(1-\tau_t^d)\lambda_t.
$$

The first-order condition with respect to $\tilde h_t$ is
$$
\frac{\partial \mathcal{L}}{\partial \tilde h_t} = 0
\quad\Rightarrow\quad
-\psi \tilde c_t^{1-\sigma}(1-\tilde h_t)^{\psi(1-\sigma)-1}
+
\lambda_t(1-\tau_t^h)\tilde w_t
=
0.
$$

Equivalently,
$$
\lambda_t(1-\tau_t^h)\tilde w_t
=
\psi \tilde c_t^{1-\sigma}(1-\tilde h_t)^{\psi(1-\sigma)-1}.
$$

Substituting the expression for $\lambda_t$ from the FOC for $\tilde c_t$, we obtain the intratemporal condition
$$
\frac{\psi \tilde c_t}{1-\tilde h_t}
=
\frac{(1-\tau_t^h)\tilde w_t}{1+\tau_t^c}.
$$

The first-order condition with respect to $\tilde k_{t+1}$ is
$$
\frac{\partial \mathcal L}{\partial \tilde k_{t+1}}=0
$$
implies
$$
-G\mu_t
+
\tilde\beta E_t\Bigg[
\lambda_{t+1}
\Big(
\tilde r_{t+1}
-
\tau_{t+1}^p(\tilde r_{t+1}-\delta)
-
\tau_{t+1}^d\big(\tilde r_{t+1}-\tau_{t+1}^p(\tilde r_{t+1}-\delta)\big)
\Big)
+
(1-\delta)\mu_{t+1}
\Bigg]
=0.
$$

Equivalently,
$$
G\mu_t
=
\tilde\beta E_t\Bigg[
\lambda_{t+1}
\Big(
\tilde r_{t+1}
-
\tau_{t+1}^p(\tilde r_{t+1}-\delta)
-
\tau_{t+1}^d\big(\tilde r_{t+1}-\tau_{t+1}^p(\tilde r_{t+1}-\delta)\big)
\Big)
+
(1-\delta)\mu_{t+1}
\Bigg].
$$

Using the FOC with respect to $\tilde x_t$,
$$
\mu_t=(1-\tau_t^d)\lambda_t,
\qquad
\mu_{t+1}=(1-\tau_{t+1}^d)\lambda_{t+1},
$$
we obtain
$$
G(1-\tau_t^d)\lambda_t
=
\tilde\beta E_t\Bigg[
(1-\tau_{t+1}^d)\lambda_{t+1}
\Big(
1+(1-\tau_{t+1}^p)(\tilde r_{t+1}-\delta)
\Big)
\Bigg].
$$

Finally, substituting the FOC with respect to $\tilde c_t$,
$$
\lambda_t
=
\frac{\tilde c_t^{-\sigma}(1-\tilde h_t)^{\psi(1-\sigma)}}{1+\tau_t^c},
$$
gives the Euler equation
$$
G(1-\tau_t^d)
\frac{\tilde c_t^{-\sigma}(1-\tilde h_t)^{\psi(1-\sigma)}}{1+\tau_t^c}
=
\tilde\beta E_t\Bigg[
(1-\tau_{t+1}^d)
\frac{\tilde c_{t+1}^{-\sigma}(1-\tilde h_{t+1})^{\psi(1-\sigma)}}{1+\tau_{t+1}^c}
\Big(
1+(1-\tau_{t+1}^p)(\tilde r_{t+1}-\delta)
\Big)
\Bigg].
$$


Using the FOC with respect to $\tilde c_t$,
$$
\lambda_t
=
\frac{\tilde c_t^{-\sigma}(1-\tilde h_t)^{\psi(1-\sigma)}}{1+\tau_t^c},
$$
and the FOC with respect to $\tilde x_t$,
$$
\mu_t=(1-\tau_t^d)\lambda_t,
$$
we can rewrite the remaining household optimality conditions more compactly.

From the FOC with respect to $\tilde h_t$, we obtain the intratemporal labor condition
$$
\frac{\psi \tilde c_t}{1-\tilde h_t}
=
\frac{(1-\tau_t^h)\tilde w_t}{1+\tau_t^c}.
$$

From the FOC with respect to $\tilde k_{t+1}$, we obtain the Euler equation
$$
G(1-\tau_t^d)
\frac{\tilde c_t^{-\sigma}(1-\tilde h_t)^{\psi(1-\sigma)}}{1+\tau_t^c}
=
\tilde\beta E_t\Bigg[
(1-\tau_{t+1}^d)
\frac{\tilde c_{t+1}^{-\sigma}(1-\tilde h_{t+1})^{\psi(1-\sigma)}}{1+\tau_{t+1}^c}
\Big(
1+(1-\tau_{t+1}^p)(\tilde r_{t+1}-\delta)
\Big)
\Bigg],
$$
where
$$
G=(1+\gamma_n)(1+\gamma_z).
$$

At the deterministic steady state, all variables are constant and the expectation operator drops out. Hence, the steady state is characterized by the following conditions.

1. Intratemporal labor condition:
$$
\frac{\psi \tilde c^{ss}}{1-\tilde h^{ss}}
=
\frac{(1-\tau_h^{ss})\tilde w^{ss}}{1+\tau_c^{ss}}.
$$

2. Euler equation:
$$
G(1-\tau_d^{ss})
\frac{(\tilde c^{ss})^{-\sigma}(1-\tilde h^{ss})^{\psi(1-\sigma)}}{1+\tau_c^{ss}}
=
\tilde\beta
(1-\tau_d^{ss})
\frac{(\tilde c^{ss})^{-\sigma}(1-\tilde h^{ss})^{\psi(1-\sigma)}}{1+\tau_c^{ss}}
\Big(
1+(1-\tau_p^{ss})(\tilde r^{ss}-\delta)
\Big).
$$

Cancelling the common terms gives the simpler steady-state Euler condition
$$
G
=
\tilde\beta
\Big(
1+(1-\tau_p^{ss})(\tilde r^{ss}-\delta)
\Big),
$$
or equivalently
$$
\frac{G}{\tilde\beta}
=
1+(1-\tau_p^{ss})(\tilde r^{ss}-\delta).
$$

3. Resource constraint:
$$
\tilde c^{ss}+\tilde x^{ss}+\tilde g^{ss}=\tilde y^{ss}.
$$

4. Capital accumulation at steady state:
$$
\tilde x^{ss}=(G-1+\delta)\tilde k^{ss}.
$$

5. Firm-side conditions:
$$
\tilde y^{ss}=(\tilde K^{ss})^\theta (z^{ss}\tilde H^{ss})^{1-\theta},
$$
$$
\tilde r^{ss}=\theta (\tilde K^{ss})^{\theta-1}(z^{ss}\tilde H^{ss})^{1-\theta},
\qquad
\tilde w^{ss}=(1-\theta)(\tilde K^{ss})^\theta (z^{ss})^{1-\theta}(\tilde H^{ss})^{-\theta}.
$$

6. Representative-agent identities:
$$
\tilde K^{ss}=\tilde k^{ss},
\qquad
\tilde H^{ss}=\tilde h^{ss},
\qquad
\tilde G^{ss}=\tilde g^{ss}.
$$

7. Government budget constraint:
$$
\tilde G^{ss}+\tilde\kappa^{ss}
=
\tau_c^{ss}\tilde C^{ss}
+
\tau_h^{ss}\tilde w^{ss}\tilde H^{ss}
+
\tau_p^{ss}(\tilde r^{ss}\tilde K^{ss}-\delta \tilde K^{ss})
+
\tau_d^{ss}
\Bigl(
\tilde r^{ss}\tilde K^{ss}-\tilde X^{ss}-\tau_p^{ss}(\tilde r^{ss}\tilde K^{ss}-\delta \tilde K^{ss})
\Bigr).
$$


```julia
# -------------------------------------------------------
# Solve the deterministic steady state of the HW6 model
# by Newton's method using numerical Jacobians
#
# Unknowns:
#   x = [k_ss, h_ss]
#
# Steady-state conditions used in the solver:
#   (1) intratemporal labor FOC
#   (2) steady-state Euler equation
#
# After solving for (k_ss, h_ss), recover all remaining
# steady-state objects:
#   c_ss, x_ss, y_ss, r_ss, w_ss, kappa_ss,
#   and the full Lecture 2 state/control blocks
#   X1_ss, X2_ss, X3_ss, y_ss, u_ss, Xbar
#
# Input:
#   params   : NamedTuple containing model parameters
#              Required fields:
#                beta, psi, sigma, gamma_n, gamma_z,
#                theta, delta, P0, P
#   tools    : module ECON8208Tools, used here for
#              numerical_jacobian
#   x0       : initial guess for [k_ss, h_ss]
#              default = [2.0, 0.30]
#   tol      : convergence tolerance for the sup-norm
#              of the residual vector
#   max_iter : maximum number of Newton iterations
#   hstep    : step size for numerical Jacobian
#   verbose  : if true, print iteration progress
#
# Output:
#   NamedTuple containing:
#     converged   : convergence flag
#     iterations  : number of Newton iterations used
#     residual    : residual vector at the solution
#     X1_ss       : steady state of individual state block
#     X2_ss       : steady state of exogenous state block
#     X3_ss       : steady state of aggregate state block
#     y_ss        : stacked known-state vector [X1_ss; X2_ss]
#     u_ss        : steady-state control vector
#     Xbar        : full stacked state vector [y_ss; X3_ss]
#     k_ss        : steady-state detrended capital
#     h_ss        : steady-state labor
#     l_ss        : steady-state leisure
#     c_ss        : steady-state detrended consumption
#     x_ss        : steady-state detrended investment
#     y_output_ss : steady-state detrended output
#     r_ss        : steady-state rental rate of capital
#     w_ss        : steady-state detrended wage
#     kappa_ss    : steady-state detrended transfer
#     z_ss        : steady-state level of detrended productivity
#     g_ss        : steady-state level of detrended government spending
#     tau_c_ss    : steady-state consumption tax rate
#     tau_h_ss    : steady-state labor income tax rate
#     tau_d_ss    : steady-state distribution tax rate
#     tau_p_ss    : steady-state profit tax rate
# -------------------------------------------------------
function solve_hw6_steady_state(params, tools;
                                x0=[2.0, 0.30],
                                tol=1e-10,
                                max_iter=200,
                                hstep=1e-6,
                                verbose=true,
                                alpha=0.5)

    beta = params.beta
    psi = params.psi
    sigma = params.sigma
    gamma_n = params.gamma_n
    gamma_z = params.gamma_z
    theta = params.theta
    delta = params.delta
    P0 = Float64.(collect(params.P0))
    P = Float64.(Matrix(params.P))

    G = (1.0 + gamma_n) * (1.0 + gamma_z)
    beta_tilde = beta * (1.0 + gamma_n) * (1.0 + gamma_z)^(1.0 - sigma)

    # ---------------------------------------------------
    # Recover the deterministic steady state of the
    # exogenous process from S̄ = P0 + P S̄
    # ---------------------------------------------------
    nS = length(P0)
    Sbar = (Matrix{Float64}(I, nS, nS) - P) \ P0

    a_ss = Sbar[1]
    tau_c_ss = Sbar[2]
    tau_h_ss = Sbar[3]
    tau_d_ss = Sbar[4]
    tau_p_ss = Sbar[5]
    logg_ss = Sbar[6]

    z_ss = exp(a_ss)
    g_ss = exp(logg_ss)

    # ---------------------------------------------------
    # Steady-state residual system in x = [k_ss, h_ss]
    # eq1 : intratemporal labor FOC
    # eq2 : steady-state Euler equation
    # ---------------------------------------------------
    function ss_residuals(x)
        k = x[1]
        h = x[2]

        if k <= 0.0 || h <= 0.0 || h >= 1.0
            return [1e8, 1e8]
        end

        K = k
        H = h

        y = K^theta * (z_ss * H)^(1.0 - theta)
        r = theta * K^(theta - 1.0) * (z_ss * H)^(1.0 - theta)
        w = (1.0 - theta) * K^theta * z_ss^(1.0 - theta) * H^(-theta)

        x_ss = (G - 1.0 + delta) * k
        c_ss = y - x_ss - g_ss

        if c_ss <= 0.0
            return [1e8, 1e8]
        end

        # Intratemporal labor condition
        eq1 = psi * c_ss / (1.0 - h) - (1.0 - tau_h_ss) * w / (1.0 + tau_c_ss)

        # Steady-state Euler equation
        eq2 = G - beta_tilde * (1.0 + (1.0 - tau_p_ss) * (r - delta))

        return [eq1, eq2]
    end

    x = Float64.(collect(x0))

    for iter in 1:max_iter
        F = ss_residuals(x)

        if norm(F, Inf) < tol
            k_ss = x[1]
            h_ss = x[2]

            K_ss = k_ss
            H_ss = h_ss
            Y_ss = K_ss^theta * (z_ss * H_ss)^(1.0 - theta)
            r_ss = theta * K_ss^(theta - 1.0) * (z_ss * H_ss)^(1.0 - theta)
            w_ss = (1.0 - theta) * K_ss^theta * z_ss^(1.0 - theta) * H_ss^(-theta)

            X_ss = (G - 1.0 + delta) * K_ss
            C_ss = Y_ss - X_ss - g_ss
            l_ss = 1.0 - h_ss

            kappa_ss =
                tau_c_ss * C_ss +
                tau_h_ss * w_ss * H_ss +
                tau_p_ss * (r_ss * K_ss - delta * K_ss) +
                tau_d_ss * (
                    r_ss * K_ss - X_ss -
                    tau_p_ss * (r_ss * K_ss - delta * K_ss)
                ) -
                g_ss

            X1_ss = [k_ss]
            X2_ss = Sbar
            X3_ss = [K_ss, H_ss, K_ss]
            y_ss = [X1_ss; X2_ss]
            u_ss = [k_ss, h_ss]
            Xbar = [y_ss; X3_ss]

            return (
                converged=true,
                iterations=iter,
                residual=F,
                X1_ss=X1_ss,
                X2_ss=X2_ss,
                X3_ss=X3_ss,
                y_ss=y_ss,
                u_ss=u_ss,
                Xbar=Xbar,
                k_ss=k_ss,
                h_ss=h_ss,
                l_ss=l_ss,
                c_ss=C_ss,
                x_ss=X_ss,
                y_output_ss=Y_ss,
                r_ss=r_ss,
                w_ss=w_ss,
                kappa_ss=kappa_ss,
                z_ss=z_ss,
                g_ss=g_ss,
                tau_c_ss=tau_c_ss,
                tau_h_ss=tau_h_ss,
                tau_d_ss=tau_d_ss,
                tau_p_ss=tau_p_ss
            )
        end

        J = tools.numerical_jacobian(ss_residuals, x; h=hstep)
        dx = J \ F
        x_new = x - dx

        if verbose
            println(
                "HW6 steady-state iteration = ", iter,
                ", sup-norm residual = ", norm(F, Inf),
                ", x = ", x
            )
        end

        x = alpha * x_new + (1 - alpha) * x
    end

    error("HW6 steady-state Newton solver did not converge within max_iter.")
end
```

### Step 1: Partition the states

Following Lecture 2, we partition the states into three blocks:
$$
X_t = (X_{1t}, X_{2t}, X_{3t}),
$$
where
- $X_{1t}$ contains individual endogenous states,
- $X_{2t}$ contains exogenous fiscal and technology shocks,
- $X_{3t}$ contains aggregate equilibrium states whose law of motion must be determined in equilibrium.

The key difference from HW5 is that taxes and prices prevent us from solving a planner's problem directly. Instead, we solve the household-firm-government equilibrium after imposing market clearing.

### Step 2: Remove discounting and cross-product terms

As in Lecture 2, we transform the local approximation into an undiscounted problem without cross-product terms. Using the effective discount factor from the detrended economy, define
$$
\tilde X_t = \tilde\beta^{t/2} X_t,
\qquad
\tilde u_t = \tilde\beta^{t/2}(u_t + R^{-1}W'X_t),
$$
and the transformed matrices
$$
\tilde Q = Q - W R^{-1} W', \qquad
\tilde A = \sqrt{\tilde\beta}(A - B R^{-1}W'), \qquad
\tilde B = \sqrt{\tilde\beta} B.
$$

This transformation allows us to use the modified Riccati and Vaughan methods developed in Lecture 2.

```julia
# -------------------------------------------------------
# Define the distorted local equilibrium in detrended form
# State vector:
#   y = [k_tilde, log z_t, tau_c_t, tau_h_t, tau_d_t, tau_p_t, log g_t]
# Aggregate vector:
#   X3 = [K_tilde, H_t, K_tilde_next]
# Control vector:
#   u = [k_tilde_next, h_t]
# -------------------------------------------------------

function r_fiscal(y, X3, u)
    k = y[1]
    a = y[2]
    tau_c = y[3]
    tau_h = y[4]
    tau_d = y[5]
    tau_p = y[6]
    logg = y[7]

    K = X3[1]
    H = X3[2]
    Kp = X3[3]

    kp = u[1]
    h = u[2]

    z = exp(a)
    g = exp(logg)
    l = 1.0 - h

    if k <= 0.0 || K <= 0.0 || kp <= 0.0 || Kp <= 0.0 || h <= 0.0 || h >= 1.0 || H <= 0.0 || H >= 1.0 || l <= 0.0
        return -1.0e12
    end

    y_agg = K^theta * (z * H)^(1.0 - theta)
    r_agg = theta * K^(theta - 1.0) * (z * H)^(1.0 - theta)
    w_agg = (1.0 - theta) * K^theta * z^(1.0 - theta) * H^(-theta)

    x_agg = G * Kp - (1.0 - delta) * K
    c_agg = y_agg - x_agg - g

    if c_agg <= 0.0
        return -1.0e12
    end

    tax_revenue = tau_c * c_agg + tau_h * w_agg * H + tau_p * (r_agg * K - delta * K) + tau_d * (r_agg * K - x_agg - tau_p * (r_agg * K - delta * K))
    kappa = tax_revenue - g

    x_ind = G * kp - (1.0 - delta) * k
    c = (r_agg * k + w_agg * h + kappa - tau_c * 0.0 - tau_h * w_agg * h - tau_p * (r_agg * k - delta * k) - tau_d * (r_agg * k - x_ind - tau_p * (r_agg * k - delta * k)) - x_ind) / (1.0 + tau_c)

    if c <= 0.0
        return -1.0e12
    end

    return ((c * l^psi)^(1.0 - sigma)) / (1.0 - sigma)
end

function g_fiscal(y, X3, u)
    s = y[2:end]
    kp = u[1]
    s_next = P0 + Matrix(P_state) * s
    return [kp; s_next]
end

ybar = [k_ss; sbar]
X3bar = [k_ss, h_ss, k_ss]
ubar = [k_ss, h_ss]

Q, W, R, Ay, Az, By = tools.lq_approximation_with_aggregates(r_fiscal, g_fiscal, ybar, X3bar, ubar; h=hstep_lq)

println("Q = ")
println(Q)
println("W = ")
println(W)
println("R = ")
println(R)
```

### Step 3: Use market clearing to eliminate aggregate variables

Market clearing conditions allow us to express aggregate states as
$$
X_{3t} = \Theta [X_{1t},X_{2t}]' + \Psi u_t.
$$
After the same transformation, this becomes
$$
\tilde X_{3t} = \tilde \Theta y_t + \tilde \Psi \tilde u_t,
\qquad
y_t = [\tilde X_{1t}, \tilde X_{2t}]'.
$$

Substituting this relation into the transformed system yields the matrices
$$
\hat A = \tilde A_y + \tilde A_z \tilde \Theta, \qquad
\hat Q = \tilde Q_y + \tilde Q_z \tilde \Theta, \qquad
\hat B = \tilde B_y + \tilde A_z \tilde \Psi, \qquad
\bar A = \tilde A_y - \tilde B_y R^{-1}\tilde \Psi' \tilde Q_z'.
$$

These matrices summarize the distorted equilibrium after imposing market clearing. Once they are constructed, the remaining problem is to solve a modified linear-quadratic system for the endogenous equilibrium laws of motion.

```julia
# -------------------------------------------------------
# Market clearing in the representative-agent equilibrium:
#   K_tilde = k_tilde,
#   H_t = h_t,
#   K_tilde_next = k_tilde_next
# -------------------------------------------------------

Theta = zeros(3, length(ybar))
Theta[1, 1] = 1.0

Psi = zeros(3, length(ubar))
Psi[2, 2] = 1.0
Psi[3, 1] = 1.0

distorted_system = tools.construct_distorted_lq_system(
    Q, W, R, Ay, Az, By, Theta, Psi, beta_tilde
)

Qhat = distorted_system.Qhat
Ahat = distorted_system.Ahat
Bhat = distorted_system.Bhat
Abar = distorted_system.Abar
Btilde_y = distorted_system.B_tilde_y

println("Qhat = ")
println(Qhat)
println("Ahat = ")
println(Ahat)
println("Bhat = ")
println(Bhat)
println("Abar = ")
println(Abar)
println("Btilde_y = ")
println(Btilde_y)
```

### Step 4: Iterate on the modified Riccati equation

The value-function matrix $P$ is obtained by iterating on the modified Riccati equation
$$
P_n
=
\hat Q
+
\bar A' P_{n+1} \hat A
-
\bar A' P_{n+1} \hat B
\left(
R + \tilde B_y' P_{n+1} \hat B
\right)^{-1}
\tilde B_y' P_{n+1} \hat A.
$$

Once the fixed point $P$ is obtained, the equilibrium policy rule in transformed variables is
$$
\tilde u_t
=
-
\left(
R + \tilde B_y' P \hat B
\right)^{-1}
\tilde B_y' P \hat A \, y_t.
$$

Thus, part (a) solves the distorted equilibrium by combining a local LQ approximation with the modified Riccati iteration from Lecture 2.

## b) Solve the same equilibrium using Vaughan's method

We now solve the same distorted local equilibrium using Vaughan's method. This is the analogue of HW5(c), but with the modified matrices from Lecture 2 that account for taxes, transfers, and equilibrium prices.

From part (a), after removing discounting and cross-product terms and imposing market clearing, the relevant state vector is
$$
y_t = [\tilde X_{1t},\tilde X_{2t}]',
$$
and the distorted equilibrium is summarized by the matrices
$$
\hat A = \tilde A_y + \tilde A_z \tilde \Theta,
\qquad
\hat Q = \tilde Q_y + \tilde Q_z \tilde \Theta,
\qquad
\hat B = \tilde B_y + \tilde A_z \tilde \Psi,
\qquad
\bar A = \tilde A_y - \tilde B_y R^{-1}\tilde \Psi'\tilde Q_z'.
$$

The goal of Vaughan's method is to recover the value-function matrix $P$ and the associated equilibrium feedback rule directly from the Hamiltonian system, rather than by iterating on the modified Riccati equation.

```julia
# -------------------------------------------------------
# Solve the same distorted equilibrium using Vaughan's
# Hamiltonian method from Lecture 2
# -------------------------------------------------------

F_vaughan, P_vaughan, H_vaughan = tools.solve_vaughan_distorted(
    Qhat, Ahat, Bhat, Abar, Btilde_y, R
)

println("F from Vaughan's method = ")
println(F_vaughan)
println("P from Vaughan's method = ")
println(P_vaughan)
```

### Step 1: Write the first-order conditions and impose market clearing

Following Lecture 2, after forming the Lagrangian and taking first-order conditions with respect to $\tilde u_t$ and $\tilde X_{t+1}$, we obtain
$$
\tilde u_t = - R^{-1} \tilde B_y' \lambda_{t+1},
$$
and
$$
\lambda_{t+1} = \tilde A_y' \lambda_{t+2} + \tilde Q_y y_{t+1} + \tilde Q_z \tilde X_{3,t+1},
$$
where $\lambda_t$ is the multiplier on the state transition constraints.

Market clearing implies that the aggregate states satisfy
$$
X_{3t} = \Theta [X_{1t},X_{2t}]' + \Psi u_t.
$$
After the same transformation used in part (a), this becomes
$$
\tilde X_{3t} = \tilde \Theta y_t + \tilde \Psi \tilde u_t.
$$

Substituting this equilibrium restriction into the first-order conditions yields a closed system in $(y_t,\lambda_t)$.

### Step 2: Construct Vaughan's Hamiltonian matrix

After substituting $\tilde X_{3t} = \tilde \Theta y_t + \tilde \Psi \tilde u_t$ into the first-order conditions and collecting terms, Lecture 2 shows that we obtain the stacked system
$$
\begin{bmatrix}
y_t \\
\lambda_t
\end{bmatrix}
=
\underbrace{
\begin{bmatrix}
\hat A^{-1} & \hat A^{-1}\hat B R^{-1}\tilde B_y' \\
\hat Q\hat A^{-1} & \hat Q\hat A^{-1}\hat B R^{-1}\tilde B_y' + \bar A'
\end{bmatrix}}_{H}
\begin{bmatrix}
y_{t+1} \\
\lambda_{t+1}
\end{bmatrix}.
$$

This matrix $H$ is the distorted analogue of Vaughan's Hamiltonian matrix from HW5. When there are no distortions, it collapses to the standard Hamiltonian used in the ordinary LQ problem.

We compute the eigenvalues and eigenvectors of $H$ and partition the selected eigenvector matrix as
$$
V =
\begin{bmatrix}
V_{11} \\
V_{21}
\end{bmatrix}.
$$
Then the quadratic value-function matrix is recovered as
$$
P = V_{21} V_{11}^{-1}.
$$

### Step 3: Recover the equilibrium feedback rule

Once $P$ is obtained from the Hamiltonian decomposition, the transformed equilibrium policy rule is
$$
\tilde u_t
=
-
\left(R + \tilde B_y' P \hat B\right)^{-1}
\tilde B_y' P \hat A \, y_t.
$$

Therefore, Vaughan's method delivers the same distorted local equilibrium as part (a), but it computes the solution directly from the Hamiltonian eigenvalue problem instead of by iterating on the modified Riccati equation.

In practice, part (a) and part (b) should produce the same matrix $P$ and the same feedback rule up to numerical error.

```julia
# -------------------------------------------------------
# Compare the modified Riccati and Vaughan solutions
# -------------------------------------------------------

println("F from modified Riccati iteration = ")
println(F_Riccati)
println("F from Vaughan's method = ")
println(F_vaughan)
println("P from modified Riccati iteration = ")
println(P_Riccati)
println("P from Vaughan's method = ")
println(P_vaughan)
println("||F_vaughan - F_Riccati|| = ", norm(F_vaughan - F_Riccati))
println("||P_vaughan - P_Riccati|| = ", norm(P_vaughan - P_Riccati))
```

#### Recover the policy functions using Vaughan's feedback matrix

Once the matrix $P$ has been computed and the feedback matrix
$$
F^{(V)} = \left(R + \tilde B_y' P \hat B\right)^{-1}\tilde B_y' P \hat A
$$
is obtained, the transformed equilibrium rule is
$$
\tilde u_t = - F^{(V)} y_t.
$$

Thus, Vaughan's method gives the same local policy functions as the modified Riccati approach in part (a), but recovers them directly from the Hamiltonian matrix of the distorted equilibrium.

## Calibration

We have the following growth model with labor supply:
\begin{aligned}
\max_{\{c_t,x_t,\ell_t\}} \quad & E \sum_{t=0}^{\infty} \beta^t 
\left\{ \frac{(c_t \ell_t^{\psi})^{1-\sigma}}{1-\sigma} \right\} N_t \\
\text{s.t.} \quad 
& c_t + x_t = k_t^{\theta}\bigl((1+\gamma_z)^t z_t h_t\bigr)^{1-\theta}, \\
& N_{t+1}  k_{t+1} = \bigl[(1-\delta)k_t + x_t\bigr] N_t, \\
& \log z_t = \rho \log z_{t-1} + \epsilon_t, \qquad \epsilon_t \sim N(0,\sigma_\epsilon^2), \\
& h_t + \ell_t = 1, \\
& N_t = (1+\gamma_n)^t, \\
& k_0 \text{ given.}
\end{aligned}



We use LQ approximation to solve the model, and we want to calibrate the following parameter:

$$
\beta,\ \psi,\ \sigma,\ \gamma_n,\ \gamma_z,\ \theta,\ \delta,\ \rho,\ \sigma_\epsilon
$$

For each parameter, we choose a moment of the data to match:

1. **$\gamma_n$**: Population growth rate

We set $\gamma_n$ to match the average annual population growth rate in the data. The corresponding moment condition is
$$
\boxed{
\log(1+\gamma_n)
=
E[\log N_t - \log N_{t-1}]
=
E[\log N_t^{\mathrm{data}} - \log N_{t-1}^{\mathrm{data}}]
}
$$


2. **$\gamma_z$**: Technology growth rate

In the detrended model, we know $\tilde{k}_t$, $\tilde{c}_t$, and $\tilde{h}_t$ are stationary. 

We have:

$$
y_t = k_t^\theta \bigl((1+\gamma_z)^t z_t h_t\bigr)^{1-\theta} = (1+\gamma_z)^t \tilde{k}_t^\theta (z_t h_t)^{1-\theta}.
$$

Since the detrended variables are moving around steady state, output per worker grows around rate $\gamma_z$.  

Therefore, we calibrate $\gamma_z$ and let match the average annual growth rate of output per worker in our model match the real data:
$$
\boxed{
E\left[
\log\left(\frac{Y_t/N_t}{Y_{t-1}/N_{t-1}}\right)
\right]
=
E\left[
\log\left(\frac{Y_t^{\mathrm{data}}/N_t^{\mathrm{data}}}{Y_{t-1}^{\mathrm{data}}/N_{t-1}^{\mathrm{data}}}\right)
\right]
}
$$

3. **$\theta$**: Capital share

We set $\theta$ to match the average capital share in the data. The corresponding moment condition is
$$
\boxed{
E\left[
\frac{
\theta k_t^{\theta-1}\bigl((1+\gamma_z)^t z_t h_t\bigr)^{1-\theta} k_t
}{
y_t
}
\right]
=
E\left[\frac{r_t k_t}{y_t}\right]
=
E\left[\frac{r_t^{\mathrm{data}} k_t^{\mathrm{data}}}{y_t^{\mathrm{data}}}\right]
}
$$


4. **$\delta$**: Depreciation rate

We calibrate $\delta$ by matching the investment-capital ratio in the model to that in the data.

From the law of motion of capital, we have
$$
K_{t+1} = (1-\delta) K_t + X_t, \\
\implies
\frac{X_t}{K_t} = \frac{K_{t+1}}{K_t} - (1-\delta)
=
(1+\gamma_n)(1+\gamma_z) - (1-\delta).
$$

Therefore, the corresponding moment condition is
$$
\boxed{
E\left[\frac{X_t}{K_t}\right]
=
E\left[\frac{X_t^{\mathrm{data}}}{K_t^{\mathrm{data}}}\right]
}
$$

5. **$\sigma$**: Relative risk aversion

In our model, we do not have a moment or variable which is directly related to $\sigma$. Therefore, we just set $\sigma = 2$ as a common choice in the literature.


6. **$\psi$**: Inverse of the Frisch elasticity of labor supply

We calibrate $\psi$ to match the average labor supply in the data.

From the intratemporal first-order condition for labor in the original model, we have
$$
\frac{
(1-\theta)\,k_t^\theta \bigl((1+\gamma_z)^t z_t\bigr)^{1-\theta} h_t^{-\theta}
}{
c_t
}
=
\frac{\psi}{1-h_t}.
$$


Hence, the parameter $\psi$ governs labor supply in the model.  
Therefore, the corresponding moment condition is
$$
\boxed{
E[h_t]
=
E[h_t^{\mathrm{data}}]
}
$$


7. **$\beta$**: Discount factor

We calibrate $\beta$ by matching the capital-output ratio in the model to that in the data.

From the steady-state Euler equation, we have
$$
\frac{1}{\tilde{\beta}}
=
\frac{1-\delta+\theta \dfrac{Y_t}{K_t}}{(1+\gamma_n)(1+\gamma_z)},
$$
where
$$
\tilde{\beta} = \beta (1+\gamma_n)(1+\gamma_z)^{1-\sigma}.
$$

Rearranging, we obtain
$$
\frac{Y_t}{K_t}
=
\frac{1}{\theta}
\left(
\frac{(1+\gamma_n)(1+\gamma_z)}{\tilde{\beta}} - (1-\delta)
\right).
$$

Hence, the discount factor $\beta$ determines the capital-output ratio in the model.  
Therefore, the corresponding moment condition is
$$
\boxed{
E\left[\frac{K_t}{Y_t}\right]
=
E\left[\frac{K_t^{\mathrm{data}}}{Y_t^{\mathrm{data}}}\right]
}
$$

8. **$\rho$**: Persistence of technology shock

We calibrate $\rho$ to match the persistence of cyclical fluctuations in output per worker.

In the model, the exogenous productivity process follows
$$
\log z_t = \rho \log z_{t-1} + \epsilon_t,
\qquad
\epsilon_t \sim N(0,\sigma_\epsilon^2).
$$

Hence, the parameter $\rho$ governs the persistence of productivity shocks and therefore affects the persistence of output fluctuations in the model.

Let $\hat y_t$ denote the HP-filtered log output per worker in the model, and let $\hat y_t^{\mathrm{data}}$ denote the corresponding series in the data.  
Then the corresponding moment condition is
$$
\boxed{
\mathrm{Corr}(\hat y_t,\hat y_{t-1})
=
\mathrm{Corr}(\hat y_t^{\mathrm{data}},\hat y_{t-1}^{\mathrm{data}})
}
$$

9. **$\sigma_\epsilon$**: Volatility of technology shock

We calibrate $\sigma_\epsilon$ to match the volatility of cyclical fluctuations in output per worker.

Since $\sigma_\epsilon$ determines the size of innovations to the productivity process, it primarily affects the volatility of output fluctuations in the model.

Let $\hat y_t$ denote the HP-filtered log output per worker in the model, and let $\hat y_t^{\mathrm{data}}$ denote the corresponding series in the data.  
Then the corresponding moment condition is
$$
\boxed{
\mathrm{Std}(\hat y_t)
=
\mathrm{Std}(\hat y_t^{\mathrm{data}})
}
$$






We construct the data moments used in the calibration as follows.

### 1. Population growth

We measure population growth by
$$
g_t^N = \log N_t - \log N_{t-1}.
$$

The data moment for $\gamma_n$ is the sample average growth rate:
$$
\hat E_T[g_t^N]
=
\frac{1}{T-1}\sum_{t=2}^T
\left(\log N_t^{\mathrm{data}}-\log N_{t-1}^{\mathrm{data}}\right).
$$

### 2. Output-per-worker growth

Let real output per worker be
$$
y_t^{\mathrm{data}} = \frac{Y_t^{\mathrm{data}}}{N_t^{\mathrm{data}}}.
$$

We measure its growth rate by
$$
g_t^y = \log y_t^{\mathrm{data}} - \log y_{t-1}^{\mathrm{data}}.
$$

The data moment for $\gamma_z$ is
$$
\hat E_T[g_t^y]
=
\frac{1}{T-1}\sum_{t=2}^T
\left(\log y_t^{\mathrm{data}}-\log y_{t-1}^{\mathrm{data}}\right).
$$

### 3. Labor share and capital share

We define factor income as
$$
\text{Factor Income}_t
=
Y_t^{\mathrm{data}}
-
\text{Proprietors' Income}_t
-
\text{Taxes on Production and Imports}_t
+
\text{Subsidies}_t.
$$

Then labor share is computed as
$$
\text{Labor Share}_t
=
\frac{\text{Compensation of Employees}_t}
{\text{Factor Income}_t}.
$$

Capital share is then
$$
\text{Capital Share}_t = 1 - \text{Labor Share}_t.
$$

The data moment for $\theta$ is
$$
\hat E_T[\text{Capital Share}_t]
=
\frac{1}{T}\sum_{t=1}^T \text{Capital Share}_t.
$$

### 4. Investment-capital ratio

Let $X_t^{\mathrm{data}}$ denote gross investment and $K_t^{\mathrm{data}}$ denote private fixed assets. We compute
$$
\frac{X_t^{\mathrm{data}}}{K_t^{\mathrm{data}}}.
$$

The data moment for $\delta$ is
$$
\hat E_T\left[\frac{X_t^{\mathrm{data}}}{K_t^{\mathrm{data}}}\right]
=
\frac{1}{T}\sum_{t=1}^T \frac{X_t^{\mathrm{data}}}{K_t^{\mathrm{data}}}.
$$

### 5. Capital-output ratio

We compute
$$
\frac{K_t^{\mathrm{data}}}{Y_t^{\mathrm{data}}}.
$$

The data moment for $\beta$ is
$$
\hat E_T\left[\frac{K_t^{\mathrm{data}}}{Y_t^{\mathrm{data}}}\right]
=
\frac{1}{T}\sum_{t=1}^T \frac{K_t^{\mathrm{data}}}{Y_t^{\mathrm{data}}}.
$$

### 6. Hours and labor supply

Let
$$
H_t^{\mathrm{emp}}
$$
denote total hours worked by employees,
$$
N_t^{\mathrm{emp}}
$$
denote the number of employees, and
$$
N_t^{\mathrm{self}}
$$
denote the number of self-employed workers.

We approximate self-employed hours by assigning them the same average hours as employees:
$$
H_t^{\mathrm{self}}
=
\frac{H_t^{\mathrm{emp}}}{N_t^{\mathrm{emp}}} N_t^{\mathrm{self}}.
$$

Total hours worked are then
$$
H_t^{\mathrm{tot}}
=
H_t^{\mathrm{emp}} + H_t^{\mathrm{self}}.
$$

Total workers are
$$
N_t^{\mathrm{tot}}
=
N_t^{\mathrm{emp}} + N_t^{\mathrm{self}}.
$$

Following Prescott, we assume that each worker is endowed with 100 hours of productive time per week. Hence, annual potential hours are
$$
H_t^{\mathrm{pot}}
=
N_t^{\mathrm{tot}} \times 52 \times 100.
$$

Therefore, labor supply is measured as
$$
h_t^{\mathrm{data}}
=
\frac{H_t^{\mathrm{tot}}}{H_t^{\mathrm{pot}}},
$$
and leisure is
$$
\ell_t^{\mathrm{data}} = 1 - h_t^{\mathrm{data}}.
$$

The data moments for $\psi$ are
$$
\hat E_T[h_t^{\mathrm{data}}]
=
\frac{1}{T}\sum_{t=1}^T h_t^{\mathrm{data}},
\qquad
\hat E_T[\ell_t^{\mathrm{data}}]
=
\frac{1}{T}\sum_{t=1}^T \ell_t^{\mathrm{data}}.
$$

### 7. Cyclical output moments

To calibrate $\rho$ and $\sigma_\epsilon$, we first construct log output per worker:
$$
\log y_t^{\mathrm{data}}
=
\log\left(\frac{Y_t^{\mathrm{data}}}{N_t^{\mathrm{data}}}\right).
$$

We then apply the HP filter and denote the cyclical component by
$$
\hat y_t^{\mathrm{data}}.
$$

The persistence moment is
$$
\mathrm{Corr}(\hat y_t^{\mathrm{data}}, \hat y_{t-1}^{\mathrm{data}}),
$$
and the volatility moment is
$$
\mathrm{Std}(\hat y_t^{\mathrm{data}}).
$$

## Calibration Results

We calibrate the model using U.S. national accounts moments over the sample period starting in 1948.  
Using the moment conditions described above, we choose the following parameter values:
$$
\beta = 0.95,\qquad
\psi = 1.65,\qquad
\sigma = 2.0,\qquad
\gamma_n = 0.01,\qquad
\gamma_z = 0.02,
$$
$$
\theta = 0.368,\qquad
\delta = 0.07,\qquad
\rho = 0.95,\qquad
\sigma_\epsilon = 0.029.
$$

With this calibration, the model matches most target moments fairly well. In particular, the model is close to the data in population growth, output-per-worker growth, labor share, capital share, the investment-capital ratio, the capital-output ratio, average hours, average leisure, and output volatility.

The main moment that is not well matched is the autocorrelation of cyclical output:
$$
corr(\hat y_t, \hat y_{t-1})
$$
However, this is normal for a simple RBC model with technology shocks only. 

We do not introduce any frictions that can generate persistent fluctuations, for example, habit formation in consumption or investment adjustment costs, thus, the model cannot generate a high persistence of output fluctuations.

A better calibration of $\rho$ and $\sigma_\epsilon$ would probably require direct productivity or TFP data. However, the assignment asks us to use only national accounts moments. Under this restriction, the parameter vector above provides a reasonable fit.
